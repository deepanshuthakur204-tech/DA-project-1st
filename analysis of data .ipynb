```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df['Month']=df['InvoiceDate'].dt.to_period('M')
monthly_sales=df.groupby('Month')['Revenue'].sum()
monthly_sales.plot(figsize=(12,6))
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
snapshot_date
```

```python
df.head()
```

```python
rfm=df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x:(snapshot_date-x.max()).days,
    'InvoiceNo':'count',
    'Revenue':'sum'
})
rfm.columns=['Recency','Frequency','Monetary']
rfm.head().reset_index()
```

```python
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm['RFM_Score'] = rfm['R_score'].astype(str) + \
                   rfm['F_score'].astype(str) + \
                   rfm['M_score'].astype(str)
```

```python
def segment_customer(row):
    if row['RFM_Score'] == '444':
        return 'VIP'
    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal'
    elif row['R_score'] == 4:
        return 'Recent'
    else:
        return 'Regular'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
sns.countplot(x='Segment', data=rfm)
plt.title("Customer Segments")
plt.show()
```

```python
monthly_sales=monthly_sales.reset_index()
monthly_sales.columns=['month','revenue']
```

```python
monthly_sales["month_number"]=np.arange(len(monthly_sales))
```

```python
monthly_sales
```

```python
from sklearn.linear_model import LinearRegression
X=monthly_sales[['month_number']]
y=monthly_sales['revenue']
model=LinearRegression()
model.fit(X,y)
```

```python
monthly_sales['Predicted']=model.predict(X)
```

```python
monthly_sales
```

```python
monthly_sales['month'] = monthly_sales['month'].dt.to_timestamp()
plt.figure(figsize=(10,5))
plt.plot(monthly_sales['month'], monthly_sales['revenue'], label='Actual')
plt.plot(monthly_sales['month'], monthly_sales['Predicted'], label='Predicted')
plt.legend()
plt.show()
```
